### Ejercicio 1 - 7 min (F)
Vamos a usar un bucle `while` para mostrar números en pantalla.

Objetivo:
Mostrar los números del **1 al 5**, uno por uno.

💡 Pista:
- Necesitas una variable contador.
- El bucle debe repetirse mientras el número sea menor o igual a 5.

```python
numero=1
while(numero<=5):
    print(f"numero contado {numero}")
    numero+=1
```

Challenge 🤺 - 15 min
Definir una variable que contenga un numero y preguntar al usuario por el numero hasta que acierte.
El programa debe tener una cantidad limitada de intentos.
Si falla imprimir si el numero a adivinar es mayor o menor que el numero ingresado y debe indicar la cantidad de intentos que le quedan.
Si acierta imprimir un mensaje.

```python
intentos=5
numeroAcertado=40##
numeroCargado=-1
while (numeroAcertado!=numeroCargado and intentos>0):
    numeroCargado=int(input("pone el numero que se te antoje"))
    
    if(numeroCargado==numeroAcertado):
        print("que Genial hacertaste el numero le hiciste un gran favor a la patria y te otorgamos el titulo de Big Boss y la medalla lirio de oro")
        break
    elif (numeroCargado<numeroAcertado):
        print(f"el Numero {numeroCargado} es Menor")

    else:
        print(f"el Numero {numeroCargado} es Mayor")
    intentos-=1

    
if(intentos>0):
    print(f"te quedan {intentos} Game over")
else:
    print("Te Quedaste sin Intentos")
```
